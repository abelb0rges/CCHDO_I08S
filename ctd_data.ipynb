```python
import pandas as pd

df = pd.read_csv('ctd_data.csv')
df.head()
```

```python
surface = df.loc[df['pres'] == 1]
```

```python
%matplotlib inline

import cartopy.crs as ccrs
import matplotlib.pyplot as plt

dx = dy = 15
bbox = (min(df['lon'])-dx, max(df['lon'])+dx,
        min(df['lat'])-dy, max(df['lat']+dx)
)
fig, ax = plt.subplots(
    figsize=(9, 9),
    subplot_kw=dict(projection=ccrs.SouthPolarStereo())
)

ax.plot(surface['lon'], surface['lat'], 'r.',
        transform=ccrs.PlateCarree())
ax.set_extent(bbox, ccrs.PlateCarree())
ax.coastlines('110m');
```

```python
fig, ax = plt.subplots(
    figsize=(9, 9),
    subplot_kw=dict(projection=ccrs.SouthPolarStereo())
)

cs = ax.scatter(
    surface['lon'], surface['lat'], 40,
    surface['sal'], alpha=0.75,
    transform=ccrs.PlateCarree()
)

ax.set_extent(bbox, ccrs.PlateCarree())
ax.coastlines('110m')
fig.colorbar(cs, extend='both', shrink=0.75);
```
